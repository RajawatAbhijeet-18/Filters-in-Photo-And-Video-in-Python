```python
cap = cv2.VideoCapture(0)
while True:
    cap.isOpened()
    ret , photo = cap.read()
    nphoto = cv2.filter2D(photo , -1 , kernal )
    fphoto = cv2.cvtColor(nphoto , cv2.COLOR_RGB2BGR)
    cv2.imshow('hi',fphoto)
    if cv2.waitKey(100) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
